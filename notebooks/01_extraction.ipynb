# Notebook 1: Data Extraction & Profiling
**Purpose:** Load the raw Online Retail data and profile its shape, data types, and completeness.
**Inputs:** `data/raw/Online_Retail.csv`
**Outputs:** None (No changes made to data here)


```python
# 1. Import required libraries for data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully.")
```

```python
# 2. Load the raw dataset
# We load the file with ISO-8859-1 encoding because this dataset has special characters
# like pound signs and accented letters that the default UTF-8 encoding cannot read.
# Without this, pandas will throw a UnicodeDecodeError.
df = pd.read_csv('../data/raw/Online_Retail.csv', encoding='ISO-8859-1')
print("Data loaded successfully.")
```

```python
# 3. Print the shape of the dataset and first 10 rows
print(f"Dataset Shape: {df.shape[0]} rows and {df.shape[1]} columns")
print("\nFirst 10 rows:")
display(df.head(10))
```

```python
# 4. Print detailed information about data types and summary statistics
print("Dataset Information:")
df.info()

print("\nSummary Statistics:")
display(df.describe(include='all'))
```

```python
# 5. Print missing value count and percentage per column
missing_counts = df.isnull().sum()
missing_percentages = (missing_counts / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_counts, 'Percentage (%)': missing_percentages})
print("Missing Values by Column:")
display(missing_df)
```

```python
# 6. Print the count of exact duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Exact Duplicate Rows: {duplicate_count}")
```

```python
# 7. Print unique value counts for Country, StockCode, and CustomerID
print(f"Unique Countries: {df['Country'].nunique()}")
print(f"Unique StockCodes: {df['StockCode'].nunique()}")
print(f"Unique CustomerIDs: {df['CustomerID'].nunique()}")
```

```python
# 8. Print InvoiceDate min and max to confirm the time range
# The InvoiceDate is a string, so we convert it temporarily to find the true min/max dates
temp_dates = pd.to_datetime(df['InvoiceDate'])
print(f"Start Date: {temp_dates.min()}")
print(f"End Date: {temp_dates.max()}")
```

```python
# 9. Print how many InvoiceNos start with 'C' (cancellations)
# We fill NA values with empty string temporarily just for the string search
cancellations_count = df['InvoiceNo'].fillna('').str.startswith('C').sum()
print(f"Number of Cancellation Transactions: {cancellations_count}")
```

```python
# 10. Print top 10 countries by transaction count
print("Top 10 Countries by Transaction Count:")
display(df['Country'].value_counts().head(10))
```
